```python
#import json file with the data
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
with open('web-app/backend/data/courses.json', encoding="utf-8") as f:
    data = json.load(f)
```

```python
# Print the type of the data
print(f"Type of data: {type(data)}")

# print the keys
print(f"Keys: {data.keys()}")

# Check if the JSON is a dictionary and extract first 100 elements from 'veranstaltungen'
if isinstance(data, dict) and "veranstaltungen" in data:
    veranstaltungen = data["veranstaltungen"]
    if "veranstaltung" in veranstaltungen:
        # Get the first 100 courses (or fewer if there are not 100 items)
        first_100_courses = veranstaltungen["veranstaltung"][:100]
        print(json.dumps(first_100_courses, indent=4))
    else:
        print("No 'veranstaltung' key found.")
else:
    print("Invalid JSON structure or no 'veranstaltungen' key found.")

# Get the list of events
veranstaltungen = data["veranstaltungen"]
veranstaltung_list = veranstaltungen["veranstaltung"]

# Loop through each event in the list
# for event in veranstaltung_list:
#    print(event["name"])  # Print the name of each event
#    print(event["bezirk"])  # Print the district (bezirk) of each event

# Get the first event in the list
first_event = veranstaltung_list[0:20]  # Get the first event

# Print all events with the title "Deutsch B1.1" in a table

# Create a DataFrame from the list of events
df = pd.DataFrame(veranstaltung_list)
# Print the first 5 rows of the DataFrame
print(df.head())
# Print the shape of the DataFrame
print(f"Shape of the DataFrame: {df.shape}")
# Print the columns of the DataFrame
print(f"Columns of the DataFrame: {df.columns}")
# Print the first 5 rows of the DataFrame
print(df.head())

```

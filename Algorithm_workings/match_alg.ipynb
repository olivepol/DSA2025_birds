```python
import pandas as pd
import numpy as np

# define levenshtein similarity function
def levenshtein(a, b, ratio=True, lowercase=True):
    # optional: convert both strings to lowercase
    if lowercase:
        a = a.lower()
        b = b.lower()

    # if one string is empty, distance = length of the other
    if a == '':
        return len(b)
    if b == '':
        return len(a)

    # initialize matrix of size (len(a)+1) x (len(b)+1)
    n, m = len(a), len(b)
    lev = np.zeros((n + 1, m + 1))

    # fill in the first column and row
    for i in range(n + 1):
        lev[i][0] = i
    for j in range(m + 1):
        lev[0][j] = j

    # fill in the matrix based on insertion, deletion, substitution
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            cost = 0 if a[i - 1] == b[j - 1] else 1
            lev[i][j] = min(
                lev[i - 1][j] + 1,        # deletion
                lev[i][j - 1] + 1,        # insertion
                lev[i - 1][j - 1] + cost  # substitution
            )

    # return similarity score (higher = better match)
    return (n + m - lev[n][m]) / (n + m) if ratio else lev[n][m]

# define the top-matching function
def get_top_matches(df, user_input, location=None, top_n=5):
    # optionally filter by district if location is provided
    if location:
        df = df[df['district'].str.lower() == location.lower()]

    # create a copy so we don’t change the original dataframe
    df = df.copy()

    # apply the similarity function to the 'last_name' (title) of each course
    df['similarity'] = df['last_name'].apply(
        lambda x: levenshtein(user_input, str(x), ratio=True)
    )

    # return the top N results, sorted by similarity score
    return df.sort_values(by='similarity', ascending=False).head(top_n)
```

```python
user_query = "social science"
user_location = "Mitte"

# get best matches
top_results = get_top_matches(df, user_query, user_location, top_n=10)

# display results
print(top_results[['last_name', 'district', 'similarity']])
```


The algorithm measures how similar two pieces of text are by counting how many small edits it would take to turn one into the other. 

These edits can be:
1. Changing one letter (like changing a to e)
2. Removing a letter
3. Adding a letter

The fewer edits needed, the more similar the two words are.

- Match - Same character, same position - 0 cost

- Substitution - Different character, same position - 1 cost

- Insertion - One string has an extra character - 1 cost

- Deletion - One string lacks a character - 1 cost
